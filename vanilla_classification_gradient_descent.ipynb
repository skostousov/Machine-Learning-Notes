```python

def softmax(data, weights):
    logits = (weights @ data)
    numerator = np.apply_along_axis((lambda x: np.exp(x)), 0, logits)
    denominator = numerator.sum(axis=0)
    return numerator/denominator
```

```python
def cross_entropy(target, prediction):
    target=target
    loss = (-1/len(target))*(np.sum(target*np.log(prediction+1e-15)))
    return loss
```

```python
def gradient(target, prediction, data):
    return (data @ np.subtract(prediction, target).T)/len(target)
```

```python
def update_weights(weights, gradient, learning_rate):
    return (weights-((learning_rate*gradient).T))
```

```python
sample_target = np.matrix([[0, 0, 1, 1], [1, 0, 0, 0], [0, 1, 0, 0]])
print(sample_target.shape)
sample_data = np.matrix([[1, 1, 1, 1], [3, 4, 5, 6], [2, 7, 3, 4]])#3 data points of 3 features each
print(sample_data.shape)
```

```python
weights = np.random.rand(3, 3)
print(weights)
best_val_loss = float('inf')
best_weights = weights
epochs = 5000

data = X_train
target = y_train
val_data = X_val
val_target = y_val
# data = data_np_transposed
# target = target_np_transposed


for epoch in range(epochs):
    prediction = softmax(data, weights)
    loss = cross_entropy(target, prediction)
    gradients = gradient(target, prediction, data)
    new_weights = update_weights(weights, gradients, 0.05)
    weights = new_weights
    val_predict = softmax(val_data, weights)
    val_loss = cross_entropy(val_target, val_predict)
    if val_loss < best_val_loss:
         best_val_loss = val_loss
         best_weights = weights
    print(f"val_loss: {val_loss:.4f} \ntrain_loss: {loss:.4f}", flush=True)
    #best_weights = weights
    #print(loss, flush=True)

prediction = softmax(data, best_weights)
print(weights)
```
